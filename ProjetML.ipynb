```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import TfidfVectorizer
from IPython.display import display


df = pd.read_csv("donnees_nettoyees_sans_vide.csv")
labels_text = df['label']
df.loc[df['label'] == 'positive','label']=3
df.loc[df['label'] == 'negative','label']=2
df.loc[df['label'] == 'neutral','label']=1

text=df['text']
labels=df['label']

text=text.values
labels=labels.values

text = np.array([word.lower() for word in text])

X_train, X_test, y_train, y_test = train_test_split(text,labels, test_size=0.2,random_state=10, stratify=labels)


custom_stopwords = ["ghir", "oyassmin", "m aha", 'sma tt', "m a", "zineb", "sma",
                    "ana", "ma", "kat", 'tt', "ga", 'a', 'aha', 'm','la','f','o','la','l','li','chi','rah',
 'w','b', 'wach', 'mn', 'had', 'ta', 't', 'wa', 'd', 'ila', 'momkin', 'li', 'l', 'o', 'f', 'chi', 'w', 'had', 'rah',
 'ya', 'wa', 'b', 'i', 'machi', 't', 'ach', 'lik', 'ta', 'mn', 'ra',
                   'al','aa','de'
 'ou',
 'howa',
 'nta',
 'daba',
 'liha',
 'ba',
 'nti',
 'hanane',
 'dyal',
 'hanan']
vectorizer_tfidf = TfidfVectorizer(lowercase=True, ngram_range=(1, 3), token_pattern='(?u)\\b\\w+\\b', min_df=1, stop_words=custom_stopwords)

X_train_tfidf = vectorizer_tfidf.fit_transform(X_train)
X_test_tfidf = vectorizer_tfidf.transform(X_test)
feature_names = vectorizer_tfidf.get_feature_names_out()

print("Mots en texte après transformation TF-IDF :")
print(feature_names)

df_tfidf = pd.DataFrame(X_train_tfidf.toarray(), columns=feature_names)
df_tfidf['label'] = y_train

# Grouper par label et calculer la moyenne des scores TF-IDF pour chaque mot
pivot_table = df_tfidf.groupby('label').mean()

# Pour chaque label, obtenir les 20 mots ayant les scores TF-IDF les plus élevés
top_20_words_by_label = {
    label: pivot_table.loc[label].nlargest(5).index.tolist()
    for label in pivot_table.index
}

# Afficher le tableau croisé dynamique
print("Tableau croisé dynamique - Top 5 mots par label :")
# display(pivot_table)
```
